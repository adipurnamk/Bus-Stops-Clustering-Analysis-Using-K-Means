```python
import tabula
```

there is 4 pages in the pdf file, use for loop to extract it to list

```python
list = []
for i in range(1, 5) :
    j = str(i)
    pdf = tabula.io.read_pdf('Halte Trans Koetaradja.pdf', pages=j, stream=True, output_format='dataframe')
    list.append(pdf)
```

the result is list of list of dataframe

```python
len(list)
```

# Cleaning process

Every dataframe is using diferent table standard, so every dataframes will be treated differently

```python
# import pandas library
import pandas as pd
```

## Table from page 1

```python
# extracting table from page 1
df1 = pd.DataFrame(list[0][0])
df1.head
```

```python
# Each page represent each corridor, so we add corridor identifier after index
# Corridor 1 : City Center-Darussalam Route
df1.insert(0, 'Corridor', 1)
corr1=df1
corr1.head()
```

```python
# Extract latitude and longitude
corr1['Latitude'] = corr1.KOORDINAT.str[:8]
corr1['Longtude'] = corr1.KOORDINAT.str[10:]
corr1.head()
```

```python
# Drop unused column
corr1.drop(columns=['NO', 'NAMA', 'KOORDINAT', 'TAHUN PEMBANGUNAN'], inplace=True)

# Rename the rest
corr1.rename(columns={'Unnamed: 1' : 'Name'}, inplace=True)
corr1
```

## Table 2

```python
# extracting table
df2 = pd.DataFrame(list[1][0])
df2.head
```

```python
# Corridor 2 : City Center-Blang Bintang Route
df2.insert(0, 'Corridor', 2)
corr2 = df2
corr2.head()
```

```python
# Extract latitude and longitude
corr2['Latitude'] = corr2.KOORDINAT.str[:8]
corr2['Longtude'] = corr2.KOORDINAT.str[10:]
corr2.head()
```

```python
# Drop unused column
corr2.drop(columns=['NO', 'KOORDINAT', 'TAHUN PEMBANGUNAN'], inplace=True)

# Rename the rest
corr2.rename(columns={'NAMA' : 'Name'}, inplace=True)
corr2
```
