# Scrapping table from pdf file

```python
# pip install tabula-py
import tabula
```

there is 4 pages in the pdf file, use for loop to extract it to list

```python
list = []
for i in range(1, 5) :
    j = str(i)
    pdf = tabula.io.read_pdf('Halte Trans Koetaradja.pdf', pages=j, stream=True, output_format='dataframe')
    list.append(pdf)
```

the result is list of list of dataframe

```python
len(list)
```

# Cleaning

Every dataframe is using diferent table standard, so every dataframes will be treated differently

```python
# import pandas library
import pandas as pd
```

## Table from page 1

```python
# extracting table from page 1
corr1 = pd.DataFrame(list[0][0])
```

```python
# Each page represent each corridor, so we add corridor identifier after index
# Corridor 1 : City Center-Darussalam Route
corr1.insert(0, 'Corridor', 1)
corr1
```

```python
# Remove unused column
corr1.drop(columns=['NO', 'NAMA', 'TAHUN PEMBANGUNAN'], inplace=True)
corr1.head()

# Rename non the rest
corr1.rename(columns={'Unnamed: 1' : 'Name', 'KOORDINAT' : 'Coordinates' }, inplace=True)
corr1
```

```python
# Extract latitude and longitude from 'KOORDINAT' column
corr1['Latitude'] = corr1.Coordinates.str[:8]
corr1['Longitude'] = corr1.Coordinates.str[10:]

corr1.head()
```

```python
# After extracting latitutde and longitude we need no more 'Coordinates' column
# Remove 'Coordinates' column
corr1.drop(columns=['Coordinates'], inplace=True)
corr1
```

```python
# Check NaN value
corr1[corr1.isna().any(axis=1)]

corr1.dtypes
```

```python
# Corridor 2 : City Center-Blang Bintang Route
corr2.insert(0, 'Corridor', 2)
corr2
```

```python
# Remove unused column
corr2.drop(columns=['NO', 'TAHUN PEMBANGUNAN'], axis=1, inplace=True)

# Rename the rest
corr2.rename(columns={'NAMA' : 'Name', 'KOORDINAT' : 'Coordinates' }, inplace=True)
corr2.head()
```

## Table 3

```python
# extracting table
corr3 = pd.DataFrame(list[2][0])

# Corridor 3 : City Center-Mata Ie Route
corr3.insert(0, 'Corridor', 3)
corr3
```

```python
# Remove unused row
# the last bus stop (index=20) will be deleted, because is not yet available, listed as on-demand ("sesuai demand" in bahasa)
corr3.drop(index=[0, 1, 20], inplace=True)

# Remove unused column
corr3.drop(columns=['Unnamed: 0', 'Unnamed: 2', 'TAHUN PEMBANGUNAN'], inplace=True)

# Rename the rest
corr3.rename(columns={'Unnamed: 1' : 'Name', 'Unnamed: 3' : 'Coordinates' }, inplace=True)
corr3.head()
```

```python
# Extract latitude and longitude
corr3['Latitude'] = corr3.Coordinates.str[:8]
corr3['Longitude'] = corr3.Coordinates.str[10:]

corr3.head()
```

```python
# Remove Coordinates column
corr3.drop(columns=['Coordinates'], inplace=True)
corr3
```

## Corridor visualization

```python
# conda install folium -c conda-forge
import folium
from folium import plugins
```

```python
corridor_colors = {
    '1' : '#FF0000', 
    '2' : '#FFFF00',
    '3' : '#00FF00',
    '4' : '#0000FF'
}

# map of Aceh
latitude = 5.5528455
longitude = 95.3192908
bus_stop_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# instantiate a mark cluster object for the incidents in the dataframe
cluster = plugins.MarkerCluster().add_to(bus_stop_map)
```

```python
# loop through the dataframe and add each data point to the mark cluster
# Label every marker with it's own PostalCode
for lat, lng, label in zip(bus_stop.Latitude, bus_stop.Longitude, bus_stop.Name):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(cluster)

points = [tuple(x) for x in bus_stop.to_numpy()]
folium.PolyLine(points, color="red", weight=2.5, opacity=1).add_to(bus_stop_map)
```
