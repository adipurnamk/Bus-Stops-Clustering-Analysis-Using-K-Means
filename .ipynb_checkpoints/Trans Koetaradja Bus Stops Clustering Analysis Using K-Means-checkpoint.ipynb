# Scrapping 

The dataset was in PDF file format. Tabula library has been used to scrap it so can be process further.

```python
# conda install -c conda-forge tabula-py
import tabula
```

```python
# There is 4 pages in the pdf file, use for loop to extract into a list
#
list = []
for i in range(1, 5) :
    j = str(i)
    pdf = tabula.io.read_pdf('Halte Trans Koetaradja.pdf', pages=j, stream=True, output_format='dataframe')
    list.append(pdf)

len(list) # The result is list of list of dataframe
#
# Every list represent each table from each page
```

# Clustering

Clustering here will be using K-Means algorithm to clustering the bus stop based on discovered venues around it.

```python
# conda install -c anaconda scikit-learn

from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
# Select just venues column and assign it into X variable
#
X = venues_df.iloc[:,3:] # Select all rows and 'Arts & Entertainment' until 'Travel & Transport'
X = np.nan_to_num(X) # Replace NaN with zero and infinity with large finite numbers 

# Normalize venue counts using MinMaxScaler from Sklearn library
#
cluster_ds = MinMaxScaler().fit_transform(X)
cluster_ds[:5]
```

```python
# check for optimum k value using yellowbrick library

from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)) # estimate 1 to 12 clusters

visualizer.fit(cluster_ds)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
from sklearn.cluster import KMeans 

# Get cluster for each bus stop
#
k = 4 # Using k value from yellowbrick plot recomendation
k_means = KMeans(init = "k-means++", n_clusters = k)
k_means.fit(cluster_ds)
labels = k_means.labels_
print(labels)
```

```python
# Append labels into dataframe
#
venues_df['Cluster'] = labels
cluster_df = venues_df
cluster_df.head()
```

```python
# Bus stop's count per cluster
#
count={}
for b in range(k):
    count["Cluster {0}".format(b+1)]=cluster_df.loc[cluster_df['Cluster'] == b].shape[0]
    
print(count)
```

```python
import matplotlib.ticker as ticker

fig, axes = plt.subplots(1, k, figsize=(20, 10), sharey=True)

axes[0].set_ylabel('Count of venues', fontsize=25)

for a in range(k):
    axes[a].xaxis.set_label_position('top')
    axes[a].set_xlabel('Cluster ' + str(a+1), fontsize=25)
    axes[a].tick_params(labelsize=20)
    plt.sca(axes[a])
    plt.xticks(rotation='vertical')
    sns.boxplot(data = cluster_df.loc[cluster_df['Cluster'] == a].drop(['Cluster', 'Corridor'],1), ax=axes[a])

plt.show()
```

## Visualization

Folium library will be used for visualization with map.

### Preparation

```python
# conda install -c conda-forge geopy
```

```python
# Get center location of Banda Aceh
#
from geopy.geocoders import Nominatim

locator = Nominatim(user_agent='myGeocoder', timeout=3) # using timeout > 1 to prevent search timeout
location = locator.geocode('Banda Aceh')
print('Location : ', location)
print('Latitude : {} \nLongitude : {}'.format(location.latitude, location.longitude))
```

We also want to plot line between each bus stops, but there is one big problem here. The bus stops data that we acquired is not sorted by it's corridor's route.  
So, to minimize the random lines in Folium map, we made new distance colummn.  
Minkowski distance formula has been used between Banda Aceh coordinates and bus stops coordinates.

```python
# Make new column for each lan & lon
#
cluster_df['lat'] = cluster_df.Coordinates.str[:8].astype(float)
cluster_df['lon'] = cluster_df.Coordinates.str[10:].astype(float)

# Minkowski distance and add new 'distance' column
#
cluster_df['lat_diff'] = (cluster_df['lat'] - location.latitude)**2 # location.latitude Banda Aceh from first map
cluster_df['lon_diff'] = (cluster_df['lat'] - location.longitude)**2 # location.longitude Banda Aceh from first map
cluster_df['Distance'] = cluster_df[['lat_diff','lon_diff']].sum(axis=1).pow(1./2)
```

### Folium map

```python
# conda install folium -c conda-forge
```

```python
# Visualization using Folium map
#
import folium

# Base map
# Using latitude and longitude of Banda Aceh
# Using OpenStreetMap tiles by default

cluster_map = folium.Map(location=[location.latitude, location.longitude], zoom_start=12) 
# Marking each bus stop based on it's cluster
#
colors=['purple','blue','green','red']
# Loop through the dataframe and add each data point to the mark cluster
for corridor, name, coordinates, cluster in zip(venues_df.Corridor, venues_df.Name, venues_df.Coordinates, venues_df.Cluster):
    latlong = [float(x) for x in coordinates.split(',')]
    folium.CircleMarker(
        latlong,
        fill=True,
        fill_opacity=0.5,
        radius=5,
        color=colors[cluster]
    ).add_to(cluster_map)

        
# display map without corridor lines
cluster_map
```

### Folium map with corridor lines

Due to limited dataset, th lines here is only to show the similarity among the corridor, not to show the bus lane.

```python
# Define each corridor colors
#
corr_colors = {
    '1':'red',
    '2':'green',
    '3':'blue',
    '4':'black'
}

# Function to draw lines
#
def draw_lines(dataframe, map_object, opacity=1):
    ''' '''
    prev_coordinates = None
    prev_corridor= '1'
    for corridor, coordinates in zip(dataframe['Corridor'], dataframe['Coordinates']):
        latlong = [float(x) for x in coordinates.split(',')]
        
        #don't draw line if line number changed
        if corridor != prev_corridor:
            prev_coordinates = None
        
        if prev_coordinates != None:
            folium.PolyLine([prev_coordinates, latlong], color=corr_colors[str(corridor)], opacity=opacity).add_to(map_object)
        prev_corridor = corridor
        prev_coordinates = latlong

# Display map without corridor lines
#
draw_lines(cluster_df.sort_values(['Corridor', 'Distance']), cluster_map, opacity=1)
cluster_map
```
